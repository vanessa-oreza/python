# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
df=pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
df
```

```python
df.set_index('Pais',inplace=True)
```

```python
df
```

```python
years=list(map(str,range(1980,2014)))
years
```

```python
colombia=df.loc['Colombia',years]
colombia
```

```python
col_dict={'Año' : colombia.index.tolist(),
          'Inmigrantes': colombia.values.tolist()}
```

```python
datos_col=pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Cantidad de inmigrantes')
plt.show()
```

Por otro lado, en el caso de Brasil, los aumentos en la inmigración comenzaron después de 2005 y parecen haber alcanzado el valor más alto en 2010.

https://seaborn.pydata.org/tutorial/color_palettes.html

```python
def generar_grafico(palette):
  fig, ax=plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax=sns.barplot(data=top_10,x='Total',y=top_10.index,hue=top_10.index, orient='h',
                 palette=palette,legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980-2013',
              loc='left', fontsize=16)
  ax.set_xlabel('Número de inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
sns.set_theme(style='darkgrid')
generar_grafico('tab10')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aplicar un tema de Seaborn
sns.set_theme(style="whitegrid")  # Puedes probar con "dark", "white", "ticks", etc.

# Crear los gráficos
fig, ax = plt.subplots(figsize=(8, 4))

# Dibujar las líneas con etiquetas para la leyenda
ax.plot(df1.loc['A'], lw=2, color='blue', label='Tienda A')
ax.plot(df1.loc['B'], lw=2, color='red', label='Tienda B')
ax.plot(df1.loc['C'], lw=2, color='purple', label='Tienda C')
ax.plot(df1.loc['D'], lw=2, color='green', label='Tienda D')

# Título y etiquetas
ax.set_title('Ventas en las tiendas A, B, C y D', loc='left')
ax.set_xlabel('Mes', fontsize=13)
ax.set_ylabel('Número de ventas', fontsize=13)
ax.xaxis.set_tick_params(labelsize=11)
ax.yaxis.set_tick_params(labelsize=10)

# Leyenda
ax.legend(title='Tiendas', loc='upper left')

# Quitar bordes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Limites del eje Y
ax.set_ylim(0, 400)

# Mostrar gráfico
plt.show()
```
