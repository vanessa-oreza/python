# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
df=pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
df
```

```python
df.set_index('Pais',inplace=True)
```

```python
df
```

```python
years=list(map(str,range(1980,2014)))
years
```

```python
colombia=df.loc['Colombia',years]
colombia
```

```python
col_dict={'Año' : colombia.index.tolist(),
          'Inmigrantes': colombia.values.tolist()}
```

```python
datos_col=pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Cantidad de inmigrantes')
plt.show()
```

Análisis

Con este gráfico, podemos observar que las tendencias de inmigración en Brasil y Argentina eran bastante similares al principio del período. Sin embargo, a partir de los años 2000, los números de inmigrantes en Argentina comenzaron a aumentar y alcanzaron los valores más altos entre 2000 y 2005, antes de disminuir y alcanzar valores muy bajos en los años más recientes. Por otro lado, en el caso de Brasil, los aumentos en la inmigración comenzaron después de 2005 y parecen haber alcanzado el valor más alto en 2010.

#Creando figuras con Matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de colombianos hacia Canadá\n1980-2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número de inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Inmigración de colombianos a Canadá\n 1980-2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Cantidad de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos hacia Canadá\n 1980-2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Cantidad de inmigrantes')

plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigración de los 4 países más grandes de Sudamérica\n 1980-2013')

axs[0,0].plot(df.loc['Colombia',years])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',years])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',years])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',years])
axs[1,1].set_title('Perú')

ymin=0
ymax=7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Número de inmigrantes')
  ax.grid()


plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
#Para crear un DataFrame con estos datos, utilizando la columna "tiendas" como índice, puedes ejecutar el siguiente código:
df = pd.DataFrame(ventas_2022, index=tiendas)
df
```

```python
#Para crear los cuatro gráficos
fig, axs=plt.subplots(2,2,figsize=(20,12))

#Ajustar los espaciados entre los subgráficos
plt.subplots_adjust(hspace=0.5, wspace=0.3)

#Título superior
fig.suptitle('Ventas de cuatro tiendas en el 2022')

#Cuatro gráficos
axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Ventas en la tienda A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Ventas en la tienda B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Ventas en la tienda C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Ventas en la tienda D')

# Agregar etiquetas para los ejes X y Y

for ax in axs.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Número de ventas')

#Definir el mismo límite en Y
ymin=0
ymax=400

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)



#No mostrar el espacio de memoria
plt.show()
```
