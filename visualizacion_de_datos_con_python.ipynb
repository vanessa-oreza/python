# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
df=pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
df
```

```python
df.set_index('Pais',inplace=True)
```

```python
df
```

```python
years=list(map(str,range(1980,2014)))
years
```

```python
colombia=df.loc['Colombia',years]
colombia
```

```python
col_dict={'Año' : colombia.index.tolist(),
          'Inmigrantes': colombia.values.tolist()}
```

```python
datos_col=pd.DataFrame(col_dict)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Cantidad de inmigrantes')
plt.show()
```

```python
datos_arg=pd.DataFrame(arg_dict)
datos_arg.tail()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_br['Inmigrantes'])
plt.plot(datos_col['Año'],datos_arg['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigración de brasileños y argentinos a Canadá')
plt.xlabel('Año')
plt.ylabel('Cantidad de inmigrantes')
plt.legend()
plt.show()
```

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], years]
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.head()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

Análisis

Con este gráfico, podemos observar que las tendencias de inmigración en Brasil y Argentina eran bastante similares al principio del período. Sin embargo, a partir de los años 2000, los números de inmigrantes en Argentina comenzaron a aumentar y alcanzaron los valores más altos entre 2000 y 2005, antes de disminuir y alcanzar valores muy bajos en los años más recientes. Por otro lado, en el caso de Brasil, los aumentos en la inmigración comenzaron después de 2005 y parecen haber alcanzado el valor más alto en 2010.
