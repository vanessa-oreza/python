```python
# Transformar para el tipo datetime definiendo el formato de fecha como DD/MM/AAAA ('%d/%m/%Y')
d2['datas_combinadas_pagamento'] = pd.to_datetime(d2['datas_combinadas_pagamento'], format='%d/%m/%Y')
d2['datas_de_pagamento'] = pd.to_datetime(d2['datas_de_pagamento'], format='%d/%m/%Y')
```
